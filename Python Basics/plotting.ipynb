```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.random.random((100, 1))
y = x
```

```python
plt.plot(x,y , "+-")
plt.show()
```

```python
plt.plot(x, x**2, "+")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.power(x, 0.5)
plt.plot(x, y, "+")
plt.grid
plt.xlabel("This is X-axis")
plt.ylabel("This is Y axis")
plt.show()
```

```python
import seaborn as sns
sns.set()
```

```python
sns.lineplot(x, y)
plt.show()
```

```python
data = pd.read_csv("mobile_cleaned.csv")
data.tail()
```

```python
ax = sns.scatterplot(x="stand_by_time", y = "battery_capacity", data = data)
```

```python
ax = sns.scatterplot(x="stand_by_time", y="battery_capacity", hue="thickness", data= data)
```

```python
# we can have histogram which is called dist plot
ax = sns.distplot(data["stand_by_time"], kde=False, rug=True)
```

```python
uniform_data = np.random.rand(10, 15)
ax = sns.heatmap(uniform_data, cmap="YlGnBu")
```
